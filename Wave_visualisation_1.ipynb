# Wave visualisation-1

This notebook aims to generate standing wave patterns on a film. The edges of the film is fixed.

```python
#Importing stuff
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

#Parameters
l = 114
h = 514 #length and height of the film
p = 1 # precision

#Constains
c = 340 #speed of the wave

m = 3
n = 5 #number of nodes

kx = m*np.pi/l
ky = n*np.pi/h #wavenumbers

k = np.sqrt(kx**2+ky**2)
wave_l = 2*np.pi/k #wavelength
f = c/k #frequency
w = 2*np.pi*f #angular speed

A = 1 #amplitude
t = 1


#kx = k*np.cos(theta)
#ky = k*np.sin(theta) #wavenumbers


x = np.arange(0,l,p)
y = np.arange(0,h,p)
xx,yy = np.meshgrid(x,y) #setting up the plot

#Function
def psi(x,y):
  return A*np.sin(kx*x)*np.sin(ky*y)*np.cos(w*t)

#Contour graph
plt.xlim(0,l)
plt.ylim(0,h)

plt.contourf(xx,yy,psi(xx,yy),cmap=cm.cool)
plt.title(r'The Contour plot')

plt.colorbar()
plt.show()

#3D graph
ax=plt.axes(projection='3d')
plt.xlim(0,l)
plt.ylim(0,h)

ax.plot_surface(xx,yy,psi(xx,yy),cmap=cm.cool)
ax.view_init(30,20)

plt.title(r'The wave in 3D')

plt.show()
```
