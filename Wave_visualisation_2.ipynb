#Wave visualisation-2

This notebook aims to generate standing wave patterns on a plate. The edges of the plate are not fixed, hence an antinode.

```python
#Importing stuff
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

#Parameters
l = 114
h = 114 #length and height of the film
p = 0.1 # precision

#Constains
c = 340 #speed of the wave

m = 5
n = 10 #number of nodes

kx = m*np.pi/l
ky = n*np.pi/h #wavenumbers

k = np.sqrt(kx**2+ky**2)
wave_l = 2*np.pi/k #wavelength
f = c/k #frequency
w = 2*np.pi*f #angular speed

A = 1 #amplitude
t = 1


#kx = k*np.cos(theta)
#ky = k*np.sin(theta) #wavenumbers


x = np.arange(0,l,p)
y = np.arange(0,h,p)
xx,yy = np.meshgrid(x,y) #setting up the plot

#Function
def psi(x,y):
  return A*np.cos(kx*x)*np.cos(ky*y)+A*np.cos(ky*x)*np.cos(kx*y)

#Contour graph
plt.figure(figsize=(8,6))
plt.xlim(0,l)
plt.ylim(0,h)

# Filled contours for background
plt.contourf(xx,yy,psi(xx,yy),cmap=cm.cool,levels=50)
plt.colorbar()
plt.contour(xx,yy,psi(xx,yy),levels=[0],colors='white',linewidths=1)

plt.title(r'The Contour Plot')
plt.figtext(
    0.5,                # X position (0 = left, 1 = right)
    -0.05,               # Y position (negative moves below the plot)
    f"White lines are where nodes are. Frequency of the wave is {f:.2f} Hz",
    wrap=True,
    horizontalalignment='center',
    fontsize=10
)
plt.tight_layout()
plt.show()

#3D graph
ax=plt.axes(projection='3d')
plt.xlim(0,l)
plt.ylim(0,h)

ax.plot_surface(xx,yy,psi(xx,yy),cmap=cm.cool, alpha=0.8)
ax.view_init(30,20)

plt.title(r'The wave in 3D')

plt.show()
```
